```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score,roc_curve
from pandas_profiling import ProfileReport
from sklearn.ensemble import BaggingClassifier
```

```python
df = pd.read_csv('winequality-red.csv')
df.head()
```

```python
Splitting data 
```

```python
x = df.drop(columns=['quality'])
y = df['quality']
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x1_train
```

# Implementation Of Bagging

```python
bag_decision_tree = BaggingClassifier(DecisionTreeClassifier(),n_estimators=10)
bag_decision_tree.fit(x1_train,y1_train)
```

```python
bag_decision_tree.predict(x1_test)
```

```python
bag_decision_tree.base_estimator
```

```python
bag_decision_tree.classes_
```

```python
bag_decision_tree.score(x1_train,y1_train)
```

```python
bag_decision_tree.score(x1_test,y1_test)
```

# Random_Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=5)
rf.fit(x1_train,y1_train)
```

```python
rf.score(x1_train,y1_train)
```

```python
rf.score(x1_test,y1_test)
```

we can tune the parameters to increase the accuracy of the model

```python
rf.estimators_
```

```python
rf.estimators_[0]
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(rf.estimators_[0],filled=True)
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(rf.estimators_[1],filled=True)
```

```python
for i in range(len(rf.estimators_)):
    tree.plot_tree(rf.estimators_[i],filled=True)
```

you can see tree structure by using estimators_

```python
grid_cv = {
    'n_estimators' : range(25,200,25),
    'criterion' :['ginny','entropy'],
    'max_depth' : range(0,10,1),
    'min_samples_leaf' : range(0,10,1)
}
```

```python
grid_rf =GridSearchCV(estimator=rf ,param_grid=grid_cv, cv=10, n_jobs=-1,verbose=1)
grid_rf.fit(x1_train,y1_train)
```

```python
grid_rf.best_params_
```

```python
rf_with_gridcv = RandomForestClassifier(criterion= 'entropy',
 max_depth= 9,
 min_samples_leaf= 1,
 n_estimators= 100)
rf_with_gridcv.fit(x1_train,y1_train)
```

```python
rf_with_gridcv.score(x1_train,y1_train)
```

```python
rf_with_gridcv.score(x1_test,y1_test)
```
